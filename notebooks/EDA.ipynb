```python
train['word_count'] = train['catalog_content'].str.count(r'\b\w+\b')
```

```python
train['sentence_count'] = train['catalog_content'].str.count(r'[.!?]+')
```

```python
train['bullet_count'] = train['catalog_content'].str.count(r'Bullet Point\s*\d+:')
```

```python
train['number_count'] = train['catalog_content'].str.count(r'\d+(?:\.\d+)?')
```

```python
train["item_name"] = train["catalog_content"].str.extract(r"Item Name:\s*(.*?)(?=\nValue:|\nBullet Point|\nUnit:|$)")
```

```python
train['brand'] = train['item_name'].str.extract(r'([A-Za-z0-9\'&.-]+\s+[A-Za-z0-9\'&.-]+)')
```
