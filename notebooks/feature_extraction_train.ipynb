```python
# Считаем слова
train['word_count'] = train['catalog_content'].str.count(r'\b\w+\b')
```

```python
train['word_count'].head()
```

```python
# Считаем предложения
train['sentence_count'] = train['catalog_content'].str.count(r'[.!?]+')
```

```python
train['sentence_count'].head()
```

```python
# Считаем буллеты
train['bullet_count'] = train['catalog_content'].str.count(r'bullet point\s*\d+:')
```

```python
train['bullet_count'].head()
```

```python
# Считаем числа
train['number_count'] = train['catalog_content'].str.count(r'\d+(?:\.\d+)?')
```

```python
train['number_count'].head()
```

```python
# Название товара
train['item_name'] = train['catalog_content'].str.extract(r'item name:\s*(.*?)(?=\nvalue:|\nbullet point|\nunit:|$)')
```

```python
train['item_name'].head()
```

```python
# Название бренда
train['brand'] = train['item_name'].str.extract(r'([A-Za-z0-9\'&.-]+\s+[A-Za-z0-9\'&.-]+)')
```

```python
train['brand'].head()
```

```python
# Кол-во слов в названии
train['title_word_count'] = train['item_name'].str.count(r'\b\w+\b')
```

```python
train['title_length'] = train['item_name'].str.len()
```

```python
# Страна производства
train['made_in_country'] = train['catalog_content'].str.extract(r'made in\s+([a-z ]{2,30})').fillna('0')
```

```python
train['made_in_country'].unique()
```

```python
# Флаги стран
train['made_in_usa'] = train['made_in_country'].str.contains(r'\b(?:usa|us|united states|united states of america|america)\b').astype(int)
train['made_in_canada'] = train['made_in_country'].str.contains(r'\bcanada\b').astype(int)
train['made_in_india'] = train['made_in_country'].str.contains(r'\bindia\b').astype(int)
train['made_in_italy'] = train['made_in_country'].str.contains(r'\bitaly\b').astype(int)
train['made_in_canada'] = train['made_in_country'].str.contains(r'\bcanada\b').astype(int)
train['made_in_mexico'] = train['made_in_country'].str.contains(r'\bmexico\b').astype(int)
train['made_in_china'] = train['made_in_country'].str.contains(r'\bchina\b').astype(int)
train['made_in_turkey'] = train['made_in_country'].str.contains(r'\bturkey\b').astype(int)
train['made_in_japan'] = train['made_in_country'].str.contains(r'\bjapan\b').astype(int)
train['made_in_chile'] = train['made_in_country'].str.contains(r'\bchile\b').astype(int)
train['made_in_france'] = train['made_in_country'].str.contains(r'\bfrance\b').astype(int)
train['made_in_spain'] = train['made_in_country'].str.contains(r'\bspain\b').astype(int)
train['made_in_thailand'] = train['made_in_country'].str.contains(r'\bthailand\b').astype(int)
train['made_in_germany'] = train['made_in_country'].str.contains(r'\bgermany\b').astype(int)
train['made_in_england'] = train['made_in_country'].str.contains(r'\bengland\b').astype(int)
train['made_in_sri_lanka'] = train['made_in_country'].str.contains(r'\bsri lanka\b').astype(int)
train['made_in_indonesia'] = train['made_in_country'].str.contains(r'\bindonesia\b').astype(int)
train['made_in_australia'] = train['made_in_country'].str.contains(r'\baustralia\b').astype(int)
train['made_in_new_zealand'] = train['made_in_country'].str.contains(r'\bnew zealand\b').astype(int)
```
