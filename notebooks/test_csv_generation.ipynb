```python
import pandas as pd
import numpy as np
import re 
```

```python
test = pd.read_csv('../data/raw/test.csv')
```

```python
test['catalog_content'] = test['catalog_content'].str.lower()
```

```python
test['text_length'] = test['catalog_content'].str.len()
```

```python
# Считаем слова
test['word_count'] = test['catalog_content'].str.count(r'\b\w+\b')
```

```python
# Считаем предложения
test['sentence_count'] = test['catalog_content'].str.count(r'[.!?]+')
```

```python
# Считаем буллеты
test['bullet_count'] = test['catalog_content'].str.count(r'bullet point\s*\d+:')
```

```python
# Считаем числа
test['number_count'] = test['catalog_content'].str.count(r'\d+(?:\.\d+)?')
```

```python
# Название товара
test['item_name'] = test['catalog_content'].str.extract(r'item name:\s*(.*?)(?=\nvalue:|\nbullet point|\nunit:|$)')
```

```python
# Название бренда
test['brand'] = test['item_name'].str.extract(r'([A-Za-z0-9\'&.-]+\s+[A-Za-z0-9\'&.-]+)')
```

```python
# Кол-во слов в названии
test['title_word_count'] = test['item_name'].str.count(r'\b\w+\b')
```

```python
test['title_length'] = test['item_name'].str.len()
```

```python
# Страна производства
test['made_in_country'] = test['catalog_content'].str.extract(r'made in\s+([a-z ]{2,30})').fillna('0')
```

```python
# Флаги стран
test['made_in_usa'] = test['made_in_country'].str.contains(r'\b(?:usa|us|united states|united states of america|america)\b').astype(int)
test['made_in_canada'] = test['made_in_country'].str.contains(r'\bcanada\b').astype(int)
test['made_in_india'] = test['made_in_country'].str.contains(r'\bindia\b').astype(int)
test['made_in_italy'] = test['made_in_country'].str.contains(r'\bitaly\b').astype(int)
test['made_in_canada'] = test['made_in_country'].str.contains(r'\bcanada\b').astype(int)
test['made_in_mexico'] = test['made_in_country'].str.contains(r'\bmexico\b').astype(int)
test['made_in_china'] = test['made_in_country'].str.contains(r'\bchina\b').astype(int)
test['made_in_turkey'] = test['made_in_country'].str.contains(r'\bturkey\b').astype(int)
test['made_in_japan'] = test['made_in_country'].str.contains(r'\bjapan\b').astype(int)
test['made_in_chile'] = test['made_in_country'].str.contains(r'\bchile\b').astype(int)
test['made_in_france'] = test['made_in_country'].str.contains(r'\bfrance\b').astype(int)
test['made_in_spain'] = test['made_in_country'].str.contains(r'\bspain\b').astype(int)
test['made_in_thailand'] = test['made_in_country'].str.contains(r'\bthailand\b').astype(int)
test['made_in_germany'] = test['made_in_country'].str.contains(r'\bgermany\b').astype(int)
test['made_in_england'] = test['made_in_country'].str.contains(r'\bengland\b').astype(int)
test['made_in_sri_lanka'] = test['made_in_country'].str.contains(r'\bsri lanka\b').astype(int)
test['made_in_indonesia'] = test['made_in_country'].str.contains(r'\bindonesia\b').astype(int)
test['made_in_australia'] = test['made_in_country'].str.contains(r'\baustralia\b').astype(int)
test['made_in_new_zealand'] = test['made_in_country'].str.contains(r'\bnew zealand\b').astype(int)
```

```python
# Единица измерения
test['unit'] = test['catalog_content'].str.extract(r'unit:\s*([A-Za-z\- ]+)')
```

```python
# Наличие упаковки
test['has_pack'] = test['catalog_content'].str.contains(r'\bpack\b').astype(int)
```

```python
# Количество упаковок
test['pack_quantity'] = test['catalog_content'].str.extract(r'pack\s*of\s*(\d+)')
```

```python
test['pack_quantity'] = test['pack_quantity'].fillna(1).astype('int')
```

```python
# Наличие состава
test['has_ingredients'] = test['catalog_content'].str.contains(r'\bingredients?\b').astype(int)
```

```python
# Диетические признаки
test['organic'] = test['catalog_content'].str.contains(r'\borganic\b').astype(int)
test['gluten_free'] = test['catalog_content'].str.contains(r'gluten[\s-]*free').astype(int)
test['vegan'] = test['catalog_content'].str.contains(r'\bvegan\b').astype(int)
test['protein'] = test['catalog_content'].str.contains(r'\bprotein\b').astype(int)
test['sugar_free'] = test['catalog_content'].str.contains(r'sugar[\s-]*free').astype(int)
test['nut_free'] = test['catalog_content'].str.contains(r'nut[\s-]*free').astype(int)
test["dairy_free"] = test["catalog_content"].str.contains(r"dairy[\s-]*free").astype(int)
test["soy_free"] = test["catalog_content"].str.contains(r"soy[\s-]*free").astype(int)
test["keto"] = test["catalog_content"].str.contains(r"\bketo\b").astype(int)
test["kosher"] = test["catalog_content"].str.contains(r"\bkosher\b").astype(int)
test["halal"] = test["catalog_content"].str.contains(r"\bhalal\b").astype(int)
```

```python
brand_frequency_dict = dict(test['brand'].value_counts()/len(test['brand'].unique()))
```

```python
test['brand_frequency'] = test['brand'].replace(brand_frequency_dict)
```

```python
unit_dict = {
    'ounce': 'oz',
    'ounces': 'oz',
    'oz': 'oz',
    'gram': 'gr',
    'grams': 'gr',
    'gramm': 'gr',
    'gr': 'gr',
    'g': 'gr',
    'pound': 'lb',
    'pounds': 'lb',
    'lb': 'lb',
    'kg': 'kg',
    'fl oz': 'fl_oz',
    'fl oz ': 'fl_oz',
    'fluid ounce': 'fl_oz',
    'fluid ounces': 'fl_oz',
    'fl ounce': 'fl_oz',
    'ml': 'ml',
    'milliliter': 'ml',
    'millilitre': 'ml',
    'mililitro': 'ml',
    'liter': 'liter',
    'liters': 'liter',
    'ltr': 'liter',
    'count': 'count',
    'ct': 'count',
    'unit': 'unit',
    'units': 'unit',
    'each': 'unit',
    'piece': 'piece',
    'pack': 'pack',
    'packs': 'pack',
    'box': 'box',
    'per box': 'box',
    'bag': 'bag',
    'bottle': 'bottle',
    'bottles': 'bottle',
    'jar': 'jar',
    'can': 'can',
    'carton': 'carton',
    'per carton': 'carton',
    'case': 'case',
    'bucket': 'bucket',
    'pouch': 'pouch',
    ' ': 'None',
    '-': 'None',
    '---': 'None'
}
```

```python
test['unit'] = test['unit'].replace(unit_dict)
```

```python
# Объем товара
test['value'] = test['catalog_content'].str.extract(r'value:\s([\d.]+)').astype(float)
```

```python
test['is_food'] = test['catalog_content'].str.contains(r'food').astype(int)
```

```python
#напитки
test['is_drink'] = test['catalog_content'].str.contains(r'coffee|tea|cocoa|espresso|drink|juice|beverage').astype(int)

#витамины
test['is_health'] = test['catalog_content'].str.contains(r'protein|vitamin|supplement|whey|collagen|creatine').astype(int)

#средства ухода
test['is_beauty'] = test['catalog_content'].str.contains(r'shampoo|conditioner|soap|cream|lotion|serum').astype(int)

#уход за животными
test['is_pet'] = test['catalog_content'].str.contains(r'dog|cat|pet|puppy|kitten').astype(int)

#уход за детьми
test['is_baby'] = test['catalog_content'].str.contains(r'baby|infant|toddler').astype(int)

#средства для дома
test['is_household'] = test['catalog_content'].str.contains(r'detergent|cleaner|laundry|dishwasher').astype(int)
```

```python
#премиальность
test["is_premium"] = test["catalog_content"].str.contains(r"premium|gourmet|artisan|imported|authentic|luxury").astype(int)

#натуральность
test["is_natural"] = test["catalog_content"].str.contains(r"natural|pure|non[\s-]*gmo|no preservatives").astype(int)
```

```python
test['total_amount'] = test['value'] * test['pack_quantity'].astype(float)
```

```python
test.to_csv('../data/extracted/test_extracted.csv')
```
